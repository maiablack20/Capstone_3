## Core Marketing Analysis

```python
store_d[store_d['Territory Manager'] == 'Jim Heck']
```

```python
store_d[store_d['Territory Manager'] == 'Bo Heap']
```

```python
merged_dta = store_s.merge(store_d)
merged_dta

print("=======================================")

merged1_dta = merged_dta[merged_dta['State'] == 'Colorado'].copy()
merged1_dta
```

```python
merged2_dta = store_s.merge(store_d)
merged2_dta

print("=======================================")

merged2_dta = merged2_dta[merged2_dta['State'] == 'Massachusetts'].copy()
merged2_dta
```

```python
merged1_dta['Month'] = pd.to_datetime(merged1_dta['Transaction Date']).dt.month
merged1_dta
```

```python
merged2_dta['Month'] = pd.to_datetime(merged2_dta['Transaction Date']).dt.month
merged2_dta
```

```python
total_rev1 = merged1_dta.groupby('Month')['Sale Amount'].sum()
print(f"Colorado's Total Revenue by {total_rev1}")
```

```python
total_rev2 = merged2_dta.groupby('Month')['Sale Amount'].sum()
print(f"Massachusetts' Total Revenue by {total_rev2}")
```

```python
ss_ranking1 = pd.DataFrame(merged1_dta.groupby('Store ID')['Sale Amount'].sum())
ss1_ranking = ss_ranking1.merge(store_d, on = 'Store ID')
ss1_ranking['Top Sales Performance'] = ss1_ranking['Sale Amount'].rank(ascending = False)
ss1_ranking.sort_values('Top Sales Performance').head(5)
print(f"Top Sales Performance by Store in Colorado: {ss1_ranking.sort_values('Top Sales Performance').head(5)}")
```

```python
ss_ranking2 = pd.DataFrame(merged2_dta.groupby('Store ID')['Sale Amount'].sum())
ss2_ranking = ss_ranking2.merge(store_d, on = 'Store ID')
ss2_ranking['Top Sales Performance'] = ss2_ranking['Sale Amount'].rank(ascending = False)
ss2_ranking.sort_values('Top Sales Performance').head(5)
print(f"Top Sales Performance by Store in Massachusetts: {ss2_ranking.sort_values('Top Sales Performance').head(5)}")
```

```python
cstmr_merge1 = merged1_dta.merge(cl, left_on = 'RewardsID', right_on = 'cust_id')
cstmr1_merge = cstmr_merge1.merge(prods, on = 'Prod Num')
cstm1r_merge = cstmr1_merge.merge(prodc.drop_duplicates('CategoryID'), on = 'CategoryID')
top_cust1 = pd.DataFrame(cstm1r_merge.groupby(['cust_id', 'name', 'Category', 'RewardsID'])['Sale Amount'].sum())
top_cust1['Top Customers'] = top_cust1.rank(ascending = False)
top_cust1.sort_values('Top Customers').head(5)
print(f"Top Customers by Sales in Colorado: {top_cust1.sort_values('Top Customers').head(5)}")
```

```python
cstmr_merge2 = merged2_dta.merge(cl, left_on = 'RewardsID', right_on = 'cust_id')
cstmr2_merge = cstmr_merge2.merge(prods, on = 'Prod Num')
cstm2r_merge = cstmr2_merge.merge(prodc.drop_duplicates('CategoryID'), on = 'CategoryID')
top_cust2 = pd.DataFrame(cstm2r_merge.groupby(['cust_id', 'name', 'Category', 'RewardsID'])['Sale Amount'].sum())
top_cust2['Top Customers'] = top_cust2.rank(ascending = False)
top_cust2.sort_values('Top Customers').head(5)
print(f"Top Customers by Sales in Massachusetts: {top_cust2.sort_values('Top Customers').head(5)}")
```

```python
num_merge1 = merged1_dta.merge(prods, on = 'Prod Num')
num1_merge = num_merge1.merge(prodc.drop_duplicates(['Category', 'CategoryID']), on = 'CategoryID')
num_trans1 = pd.DataFrame(num1_merge.groupby(['Month','Category']).count())
num_trans1['Total Transactions per Month by Category'] = num_trans1['Transaction Date'].rank(ascending = False)
num_trans1.sort_values('Total Transactions per Month by Category').head(5)
print(f"Total Transactions per Month by Category in Colorado: {[num_trans1.sort_values('Total Transactions per Month by Category').head(5) [['Transaction Date']]]}")
```

```python
num_merge2 = merged2_dta.merge(prods, on = 'Prod Num')
num2_merge = num_merge2.merge(prodc.drop_duplicates(['Category', 'CategoryID']), on = 'CategoryID')
num_trans2 = pd.DataFrame(num2_merge.groupby(['Month','Category']).count())
num_trans2['Total Transactions per Month by Category'] = num_trans2['Transaction Date'].rank(ascending = False)
num_trans2.sort_values('Total Transactions per Month by Category').head(5)
print(f"Total Transactions per Month by Category in Massachusetts: {[num_trans2.sort_values('Total Transactions per Month by Category').head(5) [['Transaction Date']]]}")
```

```python
num_merge1 = merged1_dta.merge(prods, on = "Prod Num")
num1_merge = num_merge1.merge(prodc.drop_duplicates('CategoryID'), on = 'CategoryID')
num_cat1 = pd.DataFrame(num1_merge.groupby(['Month', 'Category']).sum())
num_cat1['Total Sales Revenue per Month by Category'] = num_cat1['Sale Amount'].rank(ascending = False)
num_cat1.sort_values('Total Sales Revenue per Month by Category').head(5)
print(f"Total Sales Revenue per Month by Category in Colorado: {[num_cat1.sort_values('Total Sales Revenue per Month by Category').head(5) [['Sale Amount']]]}")
```

```python
num_merge2 = merged2_dta.merge(prods, on = "Prod Num")
num2_merge = num_merge2.merge(prodc.drop_duplicates('CategoryID'), on = 'CategoryID')
num_cat2 = pd.DataFrame(num2_merge.groupby(['Month', 'Category']).sum())
num_cat2['Total Sales Revenue per Month by Category'] = num_cat2['Sale Amount'].rank(ascending = False)
num_cat2.sort_values('Total Sales Revenue per Month by Category').head(5)
print(f"Total Sales Revenue per Month by Category in Massachusetts: {[num_cat2.sort_values('Total Sales Revenue per Month by Category').head(5) [['Sale Amount']]]}")
```

### For my analysis that was conducted, one thing that stood out the most to me was the fact that technology & accessories is the highest product category in terms of revenue, which is very remarkable. Because of the high revenue in technology & accessories, which is moreso for a educational need for our customers, I feel that with customers who make up the majority of those sales, but are not members of our rewards program need to be encouraged to become a member of our rewards program. My main reasoning for stating that the rewards program should be the marketing team's main focus is because with the customers that are not members, I feel that they would greatly benefit and appreciate being awarded for making purchases, either big or small. As a consumer myself to other businesses, a rewards program makes me, as a consumer, want to come back and be awarded perks and that's whether the perks are points that build up over time, which can be used towards a purchase, a free item (or freebie), or maybe even like a discount too. Focusing on pushing the rewards program will boost our sales in technology & accessories, but it can also boost the sales of other items that haven't garnered a high enough total revenue.
### In essence, focusing on the rewards program will be a great benefit to us as a business, but also to our customers, loyal or soon to be loyal. With the right strategy, the rewards program can garner more financial gain for the company, especially since our customer base is college students, who are in need of something that brings some type of financial ease when it comes to purchasing things needed for their classes.
