# Sales Territory Analysis - EmporiUm
## Project Overview
This notebook analyze sales data for EmporiUm, a student bookstore chain selling books, tech gear, and art supplies. The goal is to compare performance across two assigned sales territories and provide marketing recommendations.

## Sales Territores Analyzed
- Territory 1: Colorado | Mananger: Jim Heck | Region: West
- Territory 2: Texas | Mananger: Jeff "Howdy" Richards | Region: South

```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# Customer List
customers = pd.read_csv('customer_list .csv' , sep='|')
customers.info()
```

```python
# Products
products = pd.read_csv('Products .csv')
products.info()
```

```python
# Product Categories
product_categories = pd.read_csv('ProductCategories .csv')
product_categories.info()
```

```python
# Store Details
store_details = pd.read_csv('StoreDetail .csv')
store_details.info()
```

```python
# Store Sales
store_sales = pd.read_csv('StoreSales .csv')
store_sales.info()
```

## Core Marketing Analysis

```python
# Question 1: Territroy manangers and store IDs/cities for each territory
colorado_stores = store_details[store_details['State'] == 'Colorado']
texas_stores = store_details[store_details['State'] == 'Texas']

print("=" * 40)
print("COLORADO STORES")
print("=" * 40)
print(colorado_stores[['Store Location', 'State', 'Store ID', 'Territory Manager']])

print("\n" + "=" * 40)
print("TEXAS STORES")
print("=" * 40)
print(texas_stores[['Store Location', 'State', 'Store ID', 'Territory Manager']])
```

```python
# Question 2: Monthly total revenue for each territory

# Merge store_sales with store_details to get state info
sales_with_state = store_sales.merge(store_details[['Store ID', 'State']], on='Store ID', how='left')

# Filter for Colorado and Texas
colorado_sales = sales_with_state[sales_with_state['State'] == 'Colorado']
texas_sales = sales_with_state[sales_with_state['State'] == 'Texas']

# Convert Transaction Date to datetime
colorado_sales = colorado_sales.copy()
texas_sales = texas_sales.copy()
colorado_sales['Transaction Date'] = pd.to_datetime(colorado_sales['Transaction Date'])
texas_sales['Transaction Date'] = pd.to_datetime(texas_sales['Transaction Date'])

# Group by month
colorado_monthly = colorado_sales.groupby(colorado_sales['Transaction Date'].dt.to_period('M'))['Sale Amount'].sum()
texas_monthly = texas_sales.groupby(texas_sales['Transaction Date'].dt.to_period('M'))['Sale Amount'].sum()

print("=" * 40)
print("COLORADO MONTHLY REVENUE")
print("=" * 40)
print(colorado_monthly)

print("\n" + "=" * 40)
print("TEXAS MONTHLY REVENUE")
print("=" * 40)
print(texas_monthly)

```

```python
# Question 3: Rank store preformance in each territory

colorado_store_sales = colorado_sales.groupby('Store ID')['Sale Amount'].sum().reset_index()
colorado_store_sales = colorado_store_sales.merge(store_details[['Store ID', 'Store Location']], on='Store ID')
colorado_store_sales = colorado_store_sales.sort_values(by='Sale Amount', ascending=False)
colorado_store_sales['Rank'] = range(1, len(colorado_store_sales) + 1)

texas_store_sales = texas_sales.groupby('Store ID')['Sale Amount'].sum().reset_index()
texas_store_sales = texas_store_sales.merge(store_details[['Store ID', 'Store Location']], on='Store ID')
texas_store_sales = texas_store_sales.sort_values(by='Sale Amount', ascending=False)
texas_store_sales['Rank'] = range(1, len(texas_store_sales) + 1)

print("=" * 40)
print("COLORADO STORE PERFORMANCE")
print("=" * 40)
print(colorado_store_sales[['Store Location', 'Sale Amount', 'Rank']])

print("\n" + "=" * 40)
print("TEXAS STORE PERFORMANCE")
print("=" * 40)
print(texas_store_sales[['Store Location', 'Sale Amount', 'Rank']])
```

```python
# Question 4: Top customers by matching cutsomers ID to rewards ID

# Merge store_sales with store_details to get state
colorado_cutsomers = colorado_sales.groupby('RewardsID')['Sale Amount'].sum().reset_index()
texas_customers = texas_sales.groupby('RewardsID')['Sale Amount'].sum().reset_index()

# Merge with cutsomer list
colorado_customers = colorado_cutsomers.merge(customers[['cust_id', 'name']],
                                              left_on='RewardsID', 
                                              right_on='cust_id',
                                              how='left')
texas_customers = texas_customers.merge(customers[['cust_id', 'name']],
                                            left_on='RewardsID', 
                                            right_on='cust_id',
                                            how='left')

# Sort anf get top 10
colorado_customers = colorado_customers.sort_values(by='Sale Amount', ascending=False).head(10)
texas_customers = texas_customers.sort_values(by='Sale Amount', ascending=False).head(10)

print("=" * 40)
print("TOP 10 COLORADO CUSTOMERS")
print("=" * 40)
print(colorado_customers[['name','RewardsID', 'Sale Amount']])

print("\n" + "=" * 40)
print("TOP 10 TEXAS CUSTOMERS")
print("=" * 40)
print(texas_customers[['name','RewardsID', 'Sale Amount']])

```

```python
# Question 5: Transasction per month by product category

# Merge sales with products to get category info
sales_with_products = store_sales.merge(products[['Prod Num', 'CategoryID']], on='Prod Num', how='left')
sales_with_products = sales_with_products.merge(product_categories[['CategoryID', 'Category']], on='CategoryID', how='left')
sales_with_products = sales_with_products.merge(store_details[['Store ID', 'State']], on='Store ID', how='left')

# Filter for our territores 
co_tx_sales = sales_with_products[sales_with_products['State'].isin(['Colorado', 'Texas'])]

# Convert date
co_tx_sales = co_tx_sales.copy()
co_tx_sales['Transaction Date'] = pd.to_datetime(co_tx_sales['Transaction Date'])
co_tx_sales['Month'] = co_tx_sales['Transaction Date'].dt.to_period('M')

# Count transactions per month per category
category_monthly = co_tx_sales.groupby(['State', 'Month', 'Category']).size().reset_index(name='Transaction')

print(category_monthly.head(20))
```

```python
# Question 6: Total sales revenue per month by category
category_reveune = co_tx_sales.groupby(['State', 'Month', 'Category'])['Sale Amount'].sum().reset_index()

print("=" * 40)
print("COLORADO - REVENUE BY CATEGORY (first 20)")
print("=" * 40)
co_revenue = category_reveune[category_reveune['State'] == 'Colorado']
print(co_revenue.head(20))

print("\n" + "=" * 40)
print("TEXAS - REVENUE BY CATEGORY (first 20)")
print("=" * 40)
tx_revenue = category_reveune[category_reveune['State'] == 'Texas']
print(tx_revenue.head(20))

```
